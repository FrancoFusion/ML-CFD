#### Velocity

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file = '24.09.Velocity_Init_Test_0001.csv'

df = pd.read_csv(csv_file, usecols=['Velocity: Magnitude (m/s)', 'Y (m)', 'Z (m)'])
df.loc[df['Velocity: Magnitude (m/s)'] > 3e8, 'Velocity: Magnitude (m/s)'] = 0
df['Y (m)'] = df['Y (m)'].round(10)
df['Z (m)'] = df['Z (m)'].round(10)
pivot = df.pivot_table(index='Z (m)', columns='Y (m)', values='Velocity: Magnitude (m/s)', fill_value=0)
matrix = pivot.values
print(matrix.shape)
plt.imshow(matrix, origin='lower', cmap='viridis')
plt.colorbar(label='Velocity (m/s)')
plt.xlabel('Y points')
plt.ylabel('Z points')
plt.title('Velocity')
plt.tight_layout()
plt.show()
```

#### Temperature

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_file = '24.09._Temp_Init_Test_0001.csv'

df = pd.read_csv(csv_file, usecols=['Temperature (K)', 'Y (m)', 'Z (m)'])
df['Y (m)'] = df['Y (m)'].round(10)
df['Z (m)'] = df['Z (m)'].round(10)
pivot = df.pivot_table(index='Z (m)', columns='Y (m)', values='Temperature (K)', fill_value=0)
matrix = pivot.values
print(matrix.shape)

non_zero = matrix[matrix > 0]
if non_zero.size > 0:
    vmin = non_zero.min()
    vmax = matrix.max()
    plt.imshow(matrix, origin='lower', cmap='seismic', vmin=vmin, vmax=vmax)
else:
    plt.imshow(matrix, origin='lower', cmap='seismic')

plt.colorbar(label='Temperature (K)')
plt.xlabel('Y Index')
plt.ylabel('Z Index')
plt.title('Temperature')
plt.tight_layout()
plt.show()
```
